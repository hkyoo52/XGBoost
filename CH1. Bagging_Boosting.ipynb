### 회귀모델

```python
df_bikes = pd.read_csv('../데이터/Chapter01/bike_rentals_cleaned.csv')
```

```python
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=52)
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
```

```python
print('rmse : {}'.format(round(rmse,2)))
```

```python
from xgboost import XGBRegressor

model = XGBRegressor()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print('rmse : {}'.format(round(rmse,2)))
```

```python
from sklearn.model_selection import cross_val_score
model = LinearRegression()
scores = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=10)
rmse = np.sqrt(-scores)
print('rmse : {}'.format(round(rmse.mean(),2)))
```

```python
from sklearn.model_selection import cross_val_score
model = XGBRegressor()
scores = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=10)
rmse = np.sqrt(-scores)
print('rmse : {}'.format(round(rmse.mean(),2)))
```

```python
# 헤더가 없는 데이터셋을 로드합니다.
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
# df_census 열 이름을 정의합니다.
df_census.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
                   'income']
df_census.head()
```

```python
df_census.info()
```

```python
# education 열을 삭제합니다.
df_census = df_census.drop(['education'], axis=1)
# get_dummies를 사용해 숫자가 아닌 열을 바꿉니다.
df_census = pd.get_dummies(df_census)

df_census.head()
```

```python
df_census = df_census.drop('income_ <=50K', axis=1)
```

```python
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression

def cross_val(classifier,num_split=10):
    score = cross_val_score(classifier,X,y,cv = num_split)
    print('평균 정확도 : {}'.format(round(score.mean(),2)))
```

```python
cross_val(LogisticRegression())
```

```python
from xgboost import XGBClassifier
cross_val(XGBClassifier())
```
